# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
sns.set(style='whitegrid', palette='vlag')
bx = data[['RetailSales', 'RetailTransfers', 'WarehouseSales', 'ItemType']].groupby('ItemType').sum()
bx.plot.bar(figsize=(15, 7))
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
data['total_sales']= data['RetailSales'] + data['WarehouseSales']
filtro = list(data.groupby('Supplier').sum().sort_values('total_sales', ascending=False).head(10).index)
data2 = data[data['Supplier'].isin(filtro)]
```

```python
sns.set(style='whitegrid')
f, ax = plt.subplots(figsize=(15, 7))

sns.set_color_codes("pastel")
sns.barplot(x='WarehouseSales', y='Supplier', data=data2, label='WarehouseSales', color='b', estimator=sum)
sns.set_color_codes("muted")
sns.barplot(x='RetailSales', y='Supplier', data=data2, label='RetailSales', color='b', estimator=sum)

ax.legend(ncol=2, loc="upper right", frameon=True)
ax.set(xlim=(0,600000), ylabel="",
       xlabel="Top 10 Supplier - Sales by Channel")
sns.despine(left=True, bottom=True)
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
sns.set(style='whitegrid', palette='deep')
f, ax2 = plt.subplots(figsize=(15, 7))

sns.lineplot(x='Month', y='RetailSales', data=data, linewidth=1.5, label="Retail Sales")
sns.lineplot(x='Month', y='RetailTransfers', data=data, linewidth=1.5, label="Retail Transfers")
sns.lineplot(x='Month', y='WarehouseSales', data=data, linewidth=1.5, label="Warehouse Sales")

ax2.legend(ncol=2, loc="upper right", frameon=True)
ax2.set(ylabel="",
       xlabel="Sales by Channel - Per Month")
sns.despine(left=True, bottom=True)
```

## 4. Plot the same information as above but as a bar chart.

```python
#data.groupby('Month')[['RetailSales', 'RetailTransfers', 'WarehouseSales']].mean().plot.bar()

f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8), sharex=True)

sns.barplot(x='Month', y='RetailSales', data=data, palette="rocket", ax=ax1)
ax1.axhline(0, color="k", clip_on=False)
ax1.set_ylabel("Retail Sales")

sns.barplot(x='Month', y='RetailTransfers', data=data, palette="vlag", ax=ax2)
ax2.axhline(0, color="k", clip_on=False)
ax2.set_ylabel("Retail Transfers")

sns.barplot(x='Month', y='WarehouseSales', data=data, palette="deep", ax=ax3)
ax3.axhline(0, color="k", clip_on=False)
ax3.set_ylabel("Warehouse Sales")
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
sns.set(style='whitegrid', palette='pastel')
f, ax3 = plt.subplots(figsize=(15, 7))

sns.lineplot(x='Month', 
             y='RetailSales', 
             hue='ItemType', 
             data=data, 
             linewidth=1.5, 
             label="Retail Sales",
             estimator=sum)

ax3.legend(ncol=3, loc="upper left", frameon=True)
ax3.set(ylabel="",
       xlabel="Retail Sales per Item Type - Per Month")
sns.despine(left=True, bottom=True)
```

## 6. Plot the same information as above but as a bar chart.

```python
sns.set(rc={'figure.figsize':(25,10)},style='whitegrid', palette='rocket')

la = sns.barplot(x='Month', 
             y='RetailSales', 
             hue='ItemType', 
             data=data, 
             estimator=sum)

la.legend(ncol=2, loc="upper left", frameon=True)
la.set(ylabel="",
       xlabel="Retail Sales per Item Type - Per Month")
la.tick_params(labelsize=13)
la.set_xlabel('Fuel Type',fontsize=18)
la.set_ylabel('Highway MPG',fontsize=18)
sns.despine(left=True, bottom=True)
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
sns.set(rc={'figure.figsize':(15,10)},style='whitegrid', palette='pastel')

scat = sns.lmplot(x='RetailSales', 
                  y='RetailTransfers', 
                  hue='ItemType', 
                  data=data)

scat.set(xlim=(-100,1200), 
        ylim=(-100,1200),
        ylabel="",
       xlabel="Retail Sales per Item Type - Per Month")
sns.despine(left=True, bottom=True)
```

## 8. Create a scatter matrix using all the numeric fields in the data set with the plot points color-coded by Item Type.

*Hint: Seaborn's pairplot may be your best option here.*

```python
sns.pairplot(hue='ItemType', data=data, palette='rocket')
```
